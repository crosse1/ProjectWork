```python
import numpy as np
import pandas as pd
import csv

# Path to your CSV file
file_path = '/content/Transfer data project - Staging.csv'

# List to hold the rows
rows_list = []

# Open the CSV file
try:
    with open(file_path, 'r', newline='', encoding='utf-8') as csvfile:
        csvreader = csv.reader(csvfile)

        # Skip the header row if there is one
        header = next(csvreader)

        # Loop through the rows and add them to the list
        for row in csvreader:
            rows_list.append(row)
except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
print(rows_list)

# Print each row in the list
row2 = []
for row in rows_list:
    row1 = []
    i = 0
    con1 = ''
    con2 = ''
    while i < len(row):
        if i == 0:
          row1.append(row[i])
          i += 1
        elif i == 1:
          while row[i+1] != '–' and i+1< len(row)-1:
            con1 += row[i] + ' '
            i += 1
          con1 += row[i]
          row1.append(con1)
          i += 2
        else:
          con2 += row[i] + ' '
          i += 1
        if i == len(row):
          row1.append(con2)

    row2.append(row1)

print(row2)

csv_file = 'output.csv'

column_titles = ['Position', 'Name', 'College']

with open(file_path, 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(column_titles)
    # Write each row from the list of lists to the CSV file
    for row in row2:
        csvwriter.writerow(row)

#print(f"CSV file '{csv_file}' created successfully.")

```
