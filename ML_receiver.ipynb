```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import numpy as np
import pandas as pd
import csv

# Path to your CSV file
file_path = '/content/Transfer data project - Difference.csv'

# List to hold the rows
rows_list = []

try:
    with open(file_path, 'r', newline='', encoding='utf-8') as csvfile:
        csvreader = csv.reader(csvfile)

        # Skip the header row if there is one


        # Loop through the rows and add them to the list
        for row in csvreader:
            rows_list.extend(row)
except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
print(rows_list)
rows_list = rows_list[43:]

print(rows_list)

def lister(index):
    xl = []
    i = 0

    while i < len(rows_list):
        if i + index < len(rows_list):
            xl.append(rows_list[i + index])
        else:
            break
        i += index + (43-index)

    return xl

print(lister(0))
# Example data
data = {
    'conffrom': lister(7),
    'secteam': lister(8),
    'targets': lister(12),
    'yards': lister(15),
    'TD': lister(17),
    'OFFg': lister(18),
    'RECg': lister(19),
    'PassRts': lister(22),
    'yac/rec': lister(30),
    'ADOT': lister(32),
    'contCtc': lister(37),
    '1std' : lister(41)# Example target variable
}

# Convert to DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Define features and target
X = df[['conffrom', 'secteam', 'targets', 'yards', 'TD', 'OFFg', 'RECg', 'PassRts', 'yac/rec', 'ADOT', 'contCtc']]
y = df['1std']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
#final_test['Predictions1'] = (string_list)
print(predictions)
```
