```python
import pandas as pd

dataset = pd.read_csv('/content/Pff History - NN training (1).csv')

x = dataset.drop(columns=["Win_"])

y = dataset["Win_"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

import tensorflow as tf
model = tf.keras.models.Sequential()

model = tf.keras.models.Sequential()  # Create a fresh model instance
model.add(tf.keras.layers.Dense(units=256, input_shape=(x_train.shape[1],), activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=500)

model.evaluate(x_test, y_test)

final_model = tf.keras.models.Sequential()
final_model.add(tf.keras.layers.Dense(units=64, activation='sigmoid', input_dim=x_train.shape[1]))
final_model.add(tf.keras.layers.Dense(units=32, activation='sigmoid'))
final_model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))  # Adjusted to match the number of columns in y

# Compiling model
final_model.compile(loss='mae', optimizer='adam', metrics=['accuracy'])

# Training final model
final_model.fit(x_train, y_train, epochs=int(500), batch_size=32)

final_model.evaluate(x_test, y_test)


new_data = pd.read_csv('/content/Pff History - TNF (1).csv')

# Make predictions using the final model
predictions = final_model.predict(new_data)

# Add predictions as a new column to the new dataset
new_data['Predictions'] = predictions

# Save the updated dataset to a new CSV file
new_data.to_csv('/content/Predictions_Output.csv', index=False)

print("Predictions have been saved to Predictions_Output.csv")
```
